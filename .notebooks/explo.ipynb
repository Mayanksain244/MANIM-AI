# Few Shot json 

```python
import pandas as pd

# Replace this with how you load your data
# Example: all_data = pd.read_csv('your_file.csv')

output_lines = []

for desc, code in zip(all_data['Reviewed Description'], all_data['Code']):
    output_lines.append('    {')
    output_lines.append('        "role": "user",')
    output_lines.append(f'        "content": "{desc}"')
    output_lines.append('    },')
    output_lines.append('    {')
    output_lines.append('        "role": "assistant",')
    output_lines.append('        "content": """' + code.strip() + '"""')
    output_lines.append('    },')

# Remove the last comma to make valid JSON
if output_lines:
    output_lines[-1] = output_lines[-1].rstrip(',')

# Write to file
with open('manim_data.json', 'w', encoding='utf-8') as f:
    f.write('[\n' + '\n'.join(output_lines) + '\n]')
```
